```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from scipy.stats import norm, probplot

from sklearn.preprocessing import MinMaxScaler, power_transform, normalize
from imblearn.under_sampling import NearMiss
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("creditcard.csv") 
```

## Exploration des données et compréhensions du dataset 

```python
#visualization d'ensemble
df.head()
```

```python
print(f"size of the dataset : {len(df)}")
```

```python
#Recherche de valeurs absentes
df.info()
```

```python
# On constate aucune valeur absente dans le dataset
def print_na(col): 
    null = df[col].isnull().sum()
    print(f"Column {col} contains : {null} na values")
    print("-"*10)
    print()

for col in df.columns: 
    print_na(col)
```

## Exploration : Echelle des données

```python
# Time et Amount ne sont pas normalisées contrairement aux V1,V2, ... V3 
df.describe()
```

### Distribution d'Amount et Time 

```python
f, ax = plt.subplots(1,2, figsize=(14,4))
ax[0].set_title("Time distribution (without standardization)")
ax[1].set_title("Amount distribution (without standardization)")
sns.distplot(a=df["Time"], ax=ax[0])
sns.distplot(a=df["Amount"], ax=ax[1], color="g")
```

## Normalisation des variables Amount et Time  

### On observe que Time et Amount ne sont pas à la même échelle que les autres variables.On va donc utiliser MinMaxScaler pour remédier à cela


```python
#Min Max scaling
scaler = MinMaxScaler()
scaled_time = scaler.fit_transform(df["Time"].values.reshape(-1,1))
scaled_amount = scaler.fit_transform(df["Amount"].values.reshape(-1,1))
#distribution

f, ax = plt.subplots(1,2, figsize=(14,4))
ax[0].set_title("Time distribution (with Min Max Scaling)")
ax[1].set_title("Amount distribution (with Min Max Scaling)")
ax[0].set_xlabel("Time")
ax[1].set_xlabel("Amount")
sns.distplot(a=scaled_time, ax=ax[0])
sns.distplot(a=scaled_amount, ax=ax[1], color="g")

#on remplace par les valeurs mises à l'échelle :
df["Time"] = scaled_time 
df["Amount"] = scaled_amount
```

## Observons maintenant à quel point notre Dataset est déséquilibré

```python
non_fraud, fraud = df["Class"].value_counts(normalize=True).values

print(f"Frequency of Non Fraud transactions : {round(non_fraud * 100,2)} % ")
print("-"*5)
print(f"Frequency of Fraud transactions : {round(fraud * 100,2)} % ")
print("-"*5)
```

```python
# Visualisons cela : 

ax = sns.countplot(data=df, x="Class")
ax.set_xlabel("Non Fraud / Fraud")
ax.set_title("Barplot Class")
```

### Conclusion de ce déséquilibre : On va devoir utiliser des méthodes de down sampling ou de upsampling pour palier ce problème. Nous y reviendrons

# Etudions la corrélation de nos variables 

```python
corr = df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(data=corr, square=True)
plt.title("Correlation Heatmap")
plt.show()
```

### Etudions l'impact sur notre dataset de l'implémentation de NearMiss (downsampling)

```python
X = df[[col for col in df.columns if col != "Class"]]
y = df["Class"]
```

```python
nm = NearMiss()
#under indique que l'on a effectué de l'undersampling
X_under, y_under = nm.fit_resample(X,y)
```

```python
non_fraud, fraud = y_under.value_counts(normalize=True).values
print(f"Size of X before Nearmiss : {len(X)}")
print(f"Size of X after implementing Nearmiss : {len(X_under)}")
print(f"Size of y after implementing Nearmiss : {len(y_under)}")
print()
print("-"*50)
print()

print(f"Non Fraud Frequency : {round(non_fraud * 100,2)} % ")
print(f"Fraud Frquency : {round(fraud * 100, 2)} % ")
print() 
print("-"*50)
```

### Etudions la corrélation sur le downsampled dataset

```python
# En observons notre nouveau dataset, il semble qu'il ne soit pas mélangé
under_df = X_under
under_df["Class"] = y_under
#on Mélange
under_df.sample(frac=1)
```

```python
#corrélation 
under_corr = under_df.corr()
f, ax = plt.subplots(1,2, figsize=(15,5))
ax[0].set_title("Correlation without undersampling")
ax[1].set_title("Correlation with undersampling (NearMiss)")
sns.heatmap(corr, ax=ax[0],cmap="YlGnBu")
sns.heatmap(under_corr, ax=ax[1],cmap="YlGnBu")
```

## Observations : 

*** ***
    - V16,V17,V18 sont fortement corrélées (carré foncé) 
    - V2,4,11 sont positivement corrélées avec Class
    - V10,V14,16,17 sont négativement corrélées avec Class 



### On récupère les 4 variables Les plus corrélées (undersample)

```python
pos_corr = under_corr.nlargest(5, columns="Class").index
#on retire class
most_corr = pos_corr.drop("Class")
```

```python
#Distribution des variables les plus corrélées

colors = ["g", "b", "c", "r"]

f, ax = plt.subplots(2,4, figsize=(22,14))
for i in range(4): 
    ax[0][i].set_title(f"Distribution plot of {most_corr[i]}")
    ax[1][i].set_title(f"Probability plot of {most_corr[i]}")
    sns.distplot(under_df[most_corr[i]], ax=ax[0][i], color=colors[i], fit=norm)
    probplot(under_df[most_corr[i]], plot=ax[1][i])
```

#### La distribution de V19 est la plus proche d'une distribution normale

```python
neg_corr = under_corr.sort_values("Class",ascending=True)[:5]
neg_corr = neg_corr.index

f, ax = plt.subplots(2,4, figsize=(22,14))
for i in range(4): 
    ax[0][i].set_title(f"Distribution plot of {neg_corr[i]}")
    ax[1][i].set_title(f"Probability plot of {neg_corr[i]}")
    sns.distplot(under_df[neg_corr[i]], ax=ax[0][i], color=colors[i], fit=norm)
    probplot(under_df[neg_corr[i]], plot=ax[1][i])
```

### Utilisons le clustering pour avoir une idée de la capacité de nos algorithmes futures à prédire une fraud ou non

```python
#PCA on nearmiss
new_df = under_df.copy()
new_df = new_df.sample(frac=1)

pca = PCA(n_components=0.98)
pca.fit(new_df)

pca.n_components_
X_pca = pca.transform(X_under)
```

```python
target_names = y.unique()
colors = ["blue", "yellow"]

plt.figure()
for i in range(len(colors)):
    plt.scatter(x=X_pca[:,0][new_df["Class"] == i], y=X_pca[:,1][new_df["Class"] == i], alpha=0.7, color=colors[i], label=target_names[i], lw=2)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.show()
```

## Il est assez compliqué d'isoler les fraudes et non fraudes
## Procédons à une élimination des outliers

```python
def remove_outliers_fraud(col, thr=1.50):
    #fraud 
    idx = under_df[under_df["Class"] == 1].index
    col_fraud = under_df[col][idx]
    q1, q3 = np.percentile(col_fraud, 25), np.percentile(col_fraud, 75)
    cutoff = (q3 - q1) * thr
    upper, lower = q3 + cutoff , q1 - cutoff
    outliers_idx = col_fraud[(col_fraud < lower) | (col_fraud > upper)].index
    print(f"{round(len(outliers_idx) / len(col_fraud) * 100,4)} % outliers in {col}")
    print("removing")
    print("-"*20)
    under_df.drop(outliers_idx, inplace=True)
    print(f"{len(outliers_idx)} removed ...")
    print()
    
```

```python
print(len(under_df))
f, ax = plt.subplots(1,4, figsize=(18,8))
f.suptitle("Distribution des variables Négativement corrélées selon la classe (NO OUTLIERS)")
for i in range(4): 
    remove_outliers_fraud(neg_corr[i])
    sns.boxplot(data=under_df, y=neg_corr[i], x="Class",orient="v", ax=ax[i])
    ax[i].set_title(f"Distribution of {neg_corr[i]}")
```

```python
#remove outliers for negatively correlated feature

f, ax = plt.subplots(1,4, figsize=(18,8))
f.suptitle("Distribution des variables positivement corrélées selon la classe (NO OUTLIERS)")
for i in range(4): 
    remove_outliers_fraud(most_corr[i])
    sns.boxplot(data=under_df, y=most_corr[i], x="Class",orient="v", ax=ax[i])
    ax[i].set_title(f"Distribution of {most_corr[i]}")
```

```python
#PCA on nearmiss
import matplotlib.patches as mpatches

blue_patch = mpatches.Patch(color='#0A0AFF', label='No Fraud')
red_patch = mpatches.Patch(color='#AF0000', label='Fraud')

X = under_df[[col for col in under_df if col != "Class"]]
y = under_df["Class"]

X_pca = PCA(n_components=0.96, random_state=42).fit_transform(X.values)
target_names = y.unique()
colors = ["blue", "yellow"]

plt.figure()
for i in range(len(colors)):
    plt.scatter(x=X_pca[:,0], y=X_pca[:,1], alpha=0.7, c=(y == 1), label=target_names[i], lw=2, cmap="coolwarm")
plt.legend(handles=[blue_patch, red_patch])
plt.show()
```
